```python
import pandas as pd
```

```python
ccodes = pd.read_csv("../data/COW-country-codes.csv")
conflicts = pd.read_csv("../data/dyadic_mid_4.03.csv")
trade = pd.read_csv("../data/Dyadic_COW_4.0.csv")
alliances = pd.read_csv("../data/alliance_v4.1_by_directed_yearly.csv")
```

```python
conflicts_col_mapper = {"disno":"Conflict_ID", "namea":"Statecode_A", "strtyr":"Start_Year", "endyear":"End_Year", "outcome":"Outcome", "rolea":"Role_A", "roleb":"Role_B", "war":"War", "year":"Year"}
```

```python
#reset index for easy mapping
ccodes = ccodes.drop_duplicates(subset="CCode").set_index("CCode")["StateAbb"]            
```

```python
#cleanup alliances

#translate ccodes to country abbreviations
alliances["ccode1"] = alliances["ccode1"].map(ccodes)
alliances["ccode2"] = alliances["ccode2"].map(ccodes)
```

```python
#cleanup trade

#translate ccodes to country abbreviations
trade["ccode1"] = trade["ccode1"].map(ccodes)
trade["ccode2"] = trade["ccode2"].map(ccodes)
```

```python
conflicts = conflicts.rename(columns=conflicts_col_mapper)
```

```python
#write to parquet file
conflicts.to_parquet("../data/clean/conflicts.parquet")
alliances.to_parquet("../data/clean/alliances.parquet")
trade.to_parquet("../data/clean/trade.parquet")
```
