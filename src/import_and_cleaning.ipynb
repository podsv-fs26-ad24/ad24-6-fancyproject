```python
import pandas as pd
```

```python
conflicts = pd.read_csv("../data/dyadic_mid_4.03.csv")
trade = pd.read_csv("../data/Dyadic_COW_4.0.csv")
alliances = pd.read_csv("../data/alliance_v4.1_by_directed_yearly.csv")
cow_iso = pd.read_csv("../data/countries.csv")  # columns: cown, iso3, country
```

```python
conflicts_cols = ["disno", "statea", "stateb", "year", "strtyr", "endyear", "outcome", "fatlev", "hihost", "rolea", "roleb", "severity", "severitya", "severityb"]
conflicts_col_mapper = {"disno":"Conflict_ID", "statea":"Statecode_A", "stateb":"Statecode_B", "year":"Year", "strtyr":"Start_Year", "endyear":"End_Year", "outcome":"Outcome", "fatlev":"Fatality_Level", "hihost":"Hostility", "rolea":"Role_A", "roleb":"Role_B", "severity":"Severity", "severitya":"Severity_A", "severityb":"Severity_B"}
```

```python
#select columns
conflicts = conflicts[conflicts_cols]       
```

```python
# rename columns
conflicts = conflicts.rename(columns=conflicts_col_mapper)
```

```python
cow_iso = cow_iso.drop_duplicates(subset=["CCode"])
```

```python
# translate from COW countrycodes to iso codes
conflicts["Statecode_A"] = conflicts["Statecode_A"].map(
    cow_iso.set_index("CCode")["iso3"]
)
conflicts["Statecode_B"] = conflicts["Statecode_B"].map(
    cow_iso.set_index("CCode")["iso3"]
)
```

```python
# drop rows with no country code (countries that dont exist anymore today)
conflicts =  conflicts.dropna()
```

```python
test = conflicts.query("Conflict_ID == 3167")
```

```python
#write to parquet file
conflicts.to_parquet("../data/clean/conflicts.parquet")
alliances.to_parquet("../data/clean/alliances.parquet")
trade.to_parquet("../data/clean/trade.parquet")
```
